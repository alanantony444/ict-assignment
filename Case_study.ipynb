#  PART 3 — Country Analysis✨

```python
# Find total medals won by each country
total_medals = medal_winners.groupby("NOC")["Medal"].count()
print(total_medals)
```

```python
# Display top 10 countries
top10 = total_medals.sort_values(ascending=False).head(10)
print(top10)
```

```python
# Create a bar chart of top 10 countries by total medals.
import matplotlib.pyplot as plt
top10.plot(kind="bar")
plt.title("Top 10 Countries by Total Olympic Medals")
plt.xlabel("Country (NOC)")
plt.ylabel("Total Medals")
plt.show()
```

# Gold Medal Analysis 7


```python
# Filter only Gold medals
gold_medals = df[df["Medal"] == "Gold"]
# Count Gold medals by country
gold_count = gold_medals.groupby("NOC")["Medal"].count()
# Top 10 countries with most Gold medals
top10_gold = gold_count.sort_values(ascending=False).head(10)
print(top10_gold)
# Plot bar chart
top10_gold.plot(kind="bar")
plt.title("Top 10 Countries with Most Gold Medals")
plt.xlabel("Country (NOC)")
plt.ylabel("Number of Gold Medals")
plt.show()
```

# PART 4 — Gender Analysis🎇


```python
# Remove rows where Medal is null
medal_winners = df.dropna(subset=["Medal"])
# Count medals by gender
gender_medals = medal_winners["Sex"].value_counts()
print(gender_medals)
# Plot bar chart
gender_medals.plot(kind="bar")
plt.title("Medals Won by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Medals")
plt.show()
```

# Q9. Gender Trend Over Years


```python
# Group by Year and Sex
gender_trend = medal_winners.groupby(["Year", "Sex"])["Medal"].count().unstack()
print(gender_trend.head())
# Plot trend
gender_trend.plot()
plt.title("Medal Trend Over Years by Gender")
plt.xlabel("Year")
plt.ylabel("Number of Medals")
plt.show()
```

# PART 5 — Sport Analysis✨


```python
# Count medals by sport
sport_medals = medal_winners.groupby("Sport")["Medal"].count()
# Get top 10 sports
top10_sports = sport_medals.sort_values(ascending=False).head(10)
print(top10_sports)
# Plot bar chart
top10_sports.plot(kind="bar")
plt.title("Top 10 Sports by Total Medals")
plt.xlabel("Sport")
plt.ylabel("Total Medals")
plt.show()
```

# Q11. Gold Medal Sport


```python
# Filter only Gold medals
gold_medals = df[df["Medal"] == "Gold"]
# Count Gold medals by sport
gold_by_sport = gold_medals.groupby("Sport")["Medal"].count()
# Find sport with highest Gold medals
top_sport = gold_by_sport.idxmax()
max_gold = gold_by_sport.max()
print("Sport with highest Gold medals:", top_sport)
print("Total Gold medals:", max_gold)
```

#  PART 6 — Year-wise Analysis🎆


```python
# Count medals per year
medals_per_year = medal_winners.groupby("Year")["Medal"].count()
print(medals_per_year)
# Plot line graph
medals_per_year.plot(kind="line")
plt.title("Total Olympic Medals Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Medals")
plt.show()
```

#  PART 9 — Mini Investigation (Open-Ended)✨


```python
# Remove rows without medals
medal_winners = df.dropna(subset=["Medal"])
# Choose country
country = medal_winners[medal_winners["NOC"] == "USA"]
# Total medals
total_medals = country["Medal"].count()
print("Total medals:", total_medals)
# Gold medals
gold_medals = country[country["Medal"] == "Gold"]["Medal"].count()
print("Gold medals:", gold_medals)
# Best sport (most medals)
best_sport = country.groupby("Sport")["Medal"].count().sort_values(ascending=False).head(1)
print("Best sport:\n", best_sport)
# Medal trend over years
trend = country.groupby("Year")["Medal"].count()
trend.plot(kind="line", title="USA Medal Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.show()
```

# Q20. Medal Ratio
