**Question 1**

```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
print(find_max(14, 29, 35))
```

Question 2

```python
def sum_list(numbers):
    total = 0
    for x in numbers:
        total += x
    return total
sample_list = [18, 12, 30, 10, 7]
print(sum_list(sample_list))
```
