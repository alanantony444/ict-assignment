# **Question 5.**

# Checked and handled missing values in the dataset

```python
# Check missing values
print(df.isnull().sum())
df.fillna(df.mean(numeric_only=True), inplace=True)
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

# **question 6**

# Check the outliers and handle outliers in performance score using Percentiles

```python
# Define percentiles
lower = df['performance_score'].quantile(0.05)
upper = df['performance_score'].quantile(0.95)
df = df[(df['performance_score'] >= lower) & (df['performance_score'] <= upper)]
print(df)
```

# **Question 7**

# Check the gender column and do one hot encoding.


```python
# One-hot encoding
df = pd.get_dummies(df, columns=['gender'])
print(df.head())
```
