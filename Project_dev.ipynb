# **Log File Analyzer** (10)

```python
logs = [
    "2026-01-06 10:30 INFO User login success user=alex",
    "2026-01-06 10:35 ERROR Disk failure user=system",
    "2026-01-06 10:40 WARNING Low memory user=alex",
    "2026-01-06 11:00 INFO File uploaded user=john",
    "2026-01-06 11:05 ERROR Network timeout user=john",
    "2026-01-06 11:10 INFO User logout user=alex"
]
severity_count = {"INFO": 0, "WARNING": 0, "ERROR": 0}
error_types = {}
user_activity = {}
def analyze_logs():
    for log in logs:
        parts = log.split()

        severity = parts[2]
        user = parts[-1].split("=")[1]

        if severity in severity_count:
            severity_count[severity] += 1

        if severity == "ERROR":
            error_message = parts[3]
            if error_message in error_types:
                error_types[error_message] += 1
            else:
                error_types[error_message] = 1

        if user in user_activity:
            user_activity[user] += 1
        else:
            user_activity[user] = 1

def show_severity_summary():
    print("\nLog Severity Summary:")
    for key, value in severity_count.items():
        print(key, ":", value)

def show_error_analysis():
    print("\nError Type Analysis:")
    for key, value in error_types.items():
        print(key, ":", value)

def show_user_activity():
    print("\nUser Activity Analysis:")
    for key, value in user_activity.items():
        print(key, ":", value)
def filter_logs():
    level = input("Enter severity level (INFO / WARNING / ERROR): ")
    print("\nFiltered Logs:")
    for log in logs:
        if level in log:
            print(log)
analyze_logs()
while True:
    print("\n----- LOG FILE ANALYZER MENU -----")
    print("1. Show Severity Summary")
    print("2. Show Error Type Analysis")
    print("3. Show User Activity")
    print("4. Filter Logs by Severity")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        show_severity_summary()
    elif choice == "2":
        show_error_analysis()
    elif choice == "3":
        show_user_activity()
    elif choice == "4":
        filter_logs()
    elif choice == "5":
        print("Exiting Log File Analyzer...")
        break
    else:
        print("Invalid choice. Try again.")
```

# **Course Registration System**(8)

```python
courses = {
    "Python": {"credits": 3, "seats": 2},
    "Math": {"credits": 4, "seats": 1}
}
students = {
    "Alice": {"max": 6, "enrolled": []},
    "Bob": {"max": 5, "enrolled": []}
}
def total_credits(student):
    total = 0
    for c in students[student]["enrolled"]:
        total += courses[c]["credits"]
    return total
def enroll(student, course):
    if course in students[student]["enrolled"]:
        print("Already enrolled")
    elif courses[course]["seats"] == 0:
        print("No seats available")
    elif total_credits(student) + courses[course]["credits"] > students[student]["max"]:
        print("Credit limit exceeded")
    else:
        students[student]["enrolled"].append(course)
        courses[course]["seats"] -= 1
        print(student, "enrolled in", course)
enroll("Alice", "Python")
enroll("Alice", "Math")
enroll("Bob", "Math")
enroll("Bob", "Python")
```
